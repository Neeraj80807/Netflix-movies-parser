```python
import pandas as pd

df1 = pd.read_csv("netflix_movies_from_0_to_1000.csv")
df2 = pd.read_csv("netflix_movies_from_1001_to_2000.csv")
df3 = pd.read_csv("netflix_movies_from_2001_to_3000.csv")
df4 = pd.read_csv("netflix_movies_from_3000_to_4000.csv")

merged_df = pd.concat([df1, df2, df3, df4], ignore_index=True)

merged_df.to_csv("merged_file.csv", index=False)
```
